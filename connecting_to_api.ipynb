```python
import requests
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
fname = './data/data.csv'
```

```python
df = pd.read_csv(fname)
y = df['Loan_Status']
df = df.drop(['Loan_Status', 'Loan_ID'], axis=1)
_, X_test, _, y_test = train_test_split(df, y, random_state=4, train_size=0.7, stratify=y)
```

```python
jsoned = X_test.to_json()
# answer = y_test[1]
```

```python
url = 'http://192.168.0.175:6789'
```

```python
request = requests.post(url=url, json=jsoned, params={'predict_proba':True})
```
